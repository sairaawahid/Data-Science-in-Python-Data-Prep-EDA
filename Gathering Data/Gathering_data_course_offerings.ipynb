## 3. Connecting to a SQL Database

```python
# Connect to a SQLite database

import sqlite3
conn = sqlite3.connect('/content/drive/MyDrive/Data Science in Pyhton by Maven Analytics/1. Data Prep & EDA/Data/maven.db')
pd.read_sql('SELECT * FROM courses', conn)
```

```python
pd.read_sql('SELECT * FROM courses WHERE course_id = 102', conn)
```

```python
# Another example of a SQL query
pd.read_sql('SELECT * FROM courses WHERE course_id LIKE "2%"', conn)
```
